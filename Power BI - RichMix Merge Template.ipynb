```python
import glob
import pandas as pd

df_list = []
for file in glob.glob(r'C:\Users\trainor-gilhamt\Documents\Colgate NL\CP NL RM Exports Default June 2022\CP*.xlsx'):
    # Tweak this to work for your actual filepaths, if needed.
    FN, Category, Channel, Retailer= file.split('.')[0].split('-')[1:]
    df = (pd.read_excel(file)
            .assign(Folder=FN, Category=Category, Channel=Channel, Retailer=Retailer))
    df_list.append(df)

df = pd.concat(df_list)
df
```

```python
col_name="Channel"
first_col = df.pop(col_name)

col_name2="Retailer"
second_col = df.pop(col_name2)

df.insert(0, col_name, first_col)
df.insert(1, col_name2, second_col)


df
```

```python
df2 = df.drop(columns=['Folder'])
df2
```

```python

first_cols = ['Retailer' , 'Product Description' , 'UPC' , 'Dist' , 'Recd', 'Category']
last_cols = [col for col in df2.columns if col not in first_cols]

df3 = df2[first_cols+last_cols]
df3
```

```python
df3.to_excel(r'C:\Users\trainor-gilhamt\Documents\Colgate NL\Power BI\RM_MERGED.xlsx', index=False)
```
