```python
import glob
import pandas as pd

df_list = []
for file in glob.glob(r'C:\Users\ibrahimad\OneDrive - Kantar\Documents\Colgate\Colgate France\Power BI\Raw Data\CP*.xlsx'):
    # Tweak this to work for your actual filepaths, if needed.
    Folder, Category, Channel, Retailer = file.split('.')[0].split('-')[1:]  
    df = (pd.read_excel(file, header = [0,1])
            .assign(Folder=Folder, Category=Category, Channel=Channel, Retailer=Retailer))
    df_list.append(df)

df = pd.concat(df_list)
df
```

```python
Header_Retailer = df.set_index([('INFO',  'LDESC'),
            ('INFO',   'ITEM'),
            ('INFO',   'COLGATE FABRICANT'),
                        ('INFO',   'COLGATE MARQUE'),
                        ('INFO',   'MERCHANDISING COLGATE'),
                        ('INFO',   'BENEFICES'),
                        ('INFO',   'TYPES'),
                        ('INFO',   'CONDITIONNEMENT CP'),
                         ('MAT YA',   'Ventes Valeur'),
                         ('MAT YA',   'Ventes Valeur Promo'),
                         ('MAT YA',    'Ventes Valeur Hors Promo'),
                         ('MAT YA',    'Ventes UC'),
                         ('MAT YA',    'Ventes UC Promo'),
                         ('MAT YA',    'Ventes UC Hors Promo'),
                         ('MAT YA',     'DV Diffusion'),
                         ('MAT TY',   'Ventes Valeur'),
                         ('MAT TY',   'Ventes Valeur Promo'),
                         ('MAT TY',    'Ventes Valeur Hors Promo'),
                         ('MAT TY',    'Ventes UC'),
                         ('MAT TY',    'Ventes UC Promo'),
                         ('MAT TY',    'Ventes UC Hors Promo'),
                         ('MAT TY',     'DV Diffusion'),

                       
                       ]).reset_index()


Header_Retailer.columns = [''.join(col).strip() for col in Header_Retailer.columns.values]

Header_Retailer2 = Header_Retailer.drop(columns=['MAT YAVentes Valeur Promo', 'MAT YAVentes UC Promo',
                                                 'MAT TYVentes Valeur Promo',  'MAT TYVentes UC Promo'])


Rename_Retailer = Header_Retailer2.rename(columns={'INFOLDESC':'Product Description','INFOITEM':'Product Code',
                                                   'INFOCOLGATE FABRICANT':'Manufacturer', 'INFOCOLGATE MARQUE':'Brand',
                                                   'INFOMERCHANDISING COLGATE':'Attribute 1', 'INFOBENEFICES':'Attribute 2',
                                                   'INFOTYPES':'Attribute 3','INFOCONDITIONNEMENT CP':'Attribute 4',
                                                   'MAT YAVentes Valeur':'Value - MAT - YA',
                                                   'MAT YAVentes Valeur Hors Promo':'Base Value - MAT - YA',
                                                   'MAT YAVentes UC':'Units - MAT - YA',               
                                                   'MAT YAVentes UC Hors Promo':'Base Units - MAT - YA',                          
                                                   'MAT YADV Diffusion':'Distribution - MAT - YA',                                  
                                                   'MAT TYVentes Valeur':'Value - MAT - TY',
                                                   'MAT TYVentes Valeur Hors Promo':'Base Value - MAT - TY',
                                                   'MAT TYVentes UC':'Units - MAT - TY',
                                                   'MAT TYVentes UC Hors Promo':'Base Units - MAT - TY',
                                                   'MAT TYDV Diffusion':'Distribution - MAT - TY'                                    
                                                   
})


Reorg_Retailer = Rename_Retailer[['Retailer', 'Product Description', 'Product Code', 'Manufacturer', 'Brand', 
                                  'Attribute 1', 'Attribute 2', 'Attribute 3', 'Attribute 4', 
'Value - MAT - YA', 'Base Value - MAT - YA', 'Units - MAT - YA', 'Base Units - MAT - YA', 'Distribution - MAT - YA',
'Value - MAT - TY', 'Base Value - MAT - TY', 'Units - MAT - TY', 'Base Units - MAT - TY', 'Distribution - MAT - TY']]


Retailer = Reorg_Retailer
Retailer
```

```python
# tt4.loc[tt4['Parent UPC'] == 'NaN', 'Parent UPC'] = tt4['Product Code'] 
# tt4


tt4["Parent UPC"].fillna(tt4["Product Code"], inplace=True)
tt4
```

```python
tt4 = tt4.drop(columns=['Child UPC'])
tt4
```

```python
tt4['Units - MAT - Chg vs YA'] = tt4['Units - MAT - TY'] - tt4['Units - MAT - YA'] 
tt4['Base Units - MAT - Chg vs YA'] = tt4['Base Units - MAT - TY'] - tt4['Base Units - MAT - YA']
tt4['Value - MAT - Chg vs YA'] = tt4['Value - MAT - TY'] - tt4['Value - MAT - YA']
tt4['Base Value - MAT - Chg vs YA'] = tt4['Base Value - MAT - TY'] - tt4['Base Value - MAT - YA']  
tt4['Distribution - MAT - Chg vs YA'] = tt4['Distribution - MAT - TY'] - tt4['Distribution - MAT - YA']  

tt4
```
