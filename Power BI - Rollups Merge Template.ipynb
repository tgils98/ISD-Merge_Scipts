```python
import glob
import pandas as pd

df_list = []
for file in glob.glob(r'C:\Users\trainor-gilhamt\Documents\Colgate NL\CP NL RM Exports Roll Up June 2022\CP*.xlsx'):
    # Tweak this to work for your actual filepaths, if needed.
    FN, Country, Category, Retailer, Date, Rollups = file.split('.')[0].split('-')[1:]  
    df = (pd.read_excel(file)
            .assign(Folder=FN, Country=Country, Category=Category,Retailer=Retailer, Date=Date, Rollups=Rollups))
    df_list.append(df)

df = pd.concat(df_list)
df
```

```python
col_name="Retailer"
first_col = df.pop(col_name)

df.insert(0, col_name, first_col)



df
```

```python
df2 = df[['Country' , 'Retailer' , 'Parent UPC' , 'Parent Product Description' , 'Child UPC' , 'Child Product Description']]
df2
```
