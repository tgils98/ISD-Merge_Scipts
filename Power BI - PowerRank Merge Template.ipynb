```python
import glob
import pandas as pd
import numpy as np

df_list = []
for file in glob.glob(r'INSERT FILE PATH HERE\ ColgateX*.xlsx'):
    # Tweak this to work for your actual filepaths, if needed.
    FN, Category, Retailer, Date = file.split('.')[0].split('-')[1:]  
    df = (pd.read_excel(file, skiprows=2)
            .assign(Folder=FN, Category=Category, Retailer=Retailer, Date=Date))
    df_list.append(df)

df = pd.concat(df_list)
df
```

```python
col_name="Channel"
first_col = df.pop(col_name)

col_name2="Retailer"
second_col = df.pop(col_name2)

df.insert(0, col_name, first_col)
df.insert(1, col_name2, second_col)


df
```

```python
#Update '' with the name of your retailers

df['Retailer'] = np.where((df['Retailer'] == 'HM') & (df['Retailer'] == 'TM') & (df['Retailer'] == 'SM'), 'PowerRank', df['Retailer'])

df
```

```python
df2 = df.drop(columns=['Folder', 'Date'])
df2
```

```python
df3 = df2.rename(columns={'CP Seg':'Attribute 1', 'CP SubSeg':'Attribute 2', 'CP Seg.1':'Attribute 1 Rank', 'CP SubSeg.1':'Attribute 2 Rank'})

df3
```

```python
df3.to_excel(r'INSERT YOUR FILE PATH HERE\PowerRank_Merged.xlsx', index=False)
```
